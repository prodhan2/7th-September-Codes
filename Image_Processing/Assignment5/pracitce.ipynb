```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```python
image = cv2.imread('sun.png' , cv2.IMREAD_GRAYSCALE)
# plt.imshow(image , cmap='gray')

kernel = np.ones((3,3) , np.float32)/9

smoothed  = cv2.filter2D(image , -1 , kernel)
# plt.imshow(smoothed , cmap='gray')

sobelx = np.array([[-1 , 0 ,1],[-2 , 0 ,2], [-1 , 0 ,1]])
sobely  = np.array([[-1 , -2 , -1] , [0 , 0 , 0] , [1 , 2 ,1]])

sobelx_image = cv2.filter2D(image  , -1 , sobelx)
sobely_image = cv2.filter2D(image  , -1 , sobely)

# plt.imshow(sobelx_image , cmap='gray')


prewitx = np.array([[-1 , 0 ,1],[-1 , 0 ,1], [-1 , 0 ,1]])
prewity  = np.array([[-1 , -1 , -1] , [0 , 0 , 0] , [1 , 1 ,1]])

prewitx_image = cv2.filter2D(image  , -1 , prewitx)
prewity_image = cv2.filter2D(image  , -1 , prewity)

# plt.imshow(prewitx_image, cmap='gray')

Laplacian =  np.array([[0 , -1 , 0 ] , [-1 , 4 , -1], [0 , -1 , 0]])


laplacian_image = cv2.filter2D(image , -1, Laplacian)

images = [image , smoothed , sobelx_image , sobely_image , prewitx_image , prewity_image , laplacian_image]
titles = ['Original Image' , 'Smoothed Image' , 'Sobel X' , 'Sobel Y' , 'Prewit X' , 'Prewit Y' , 'Laplacian']



plt.figure(figsize=(10,6))
for i in range (len(images)):
   plt.subplot(2,4, i+1)
   plt.imshow(images[i] , cmap= 'gray')
   plt.title(titles[i])

plt.tight_layout()
plt.show()

```
