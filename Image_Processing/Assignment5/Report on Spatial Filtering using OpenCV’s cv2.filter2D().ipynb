```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ======================
# Read input image directly as Grayscale
# ======================
img_gray = cv2.imread("sun.png", cv2.IMREAD_GRAYSCALE)

# ======================
# 1. Smoothing / Average Filter
# ======================
kernel_avg = np.ones((3,3), np.float32) / 9
smooth = cv2.filter2D(img_gray, -1, kernel_avg)

# ======================
# 2. Sobel Filters
# ======================
sobel_x_kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y_kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

sobel_x = cv2.filter2D(img_gray, -1, sobel_x_kernel)
sobel_y = cv2.filter2D(img_gray, -1, sobel_y_kernel)

# ======================
# 3. Prewitt Filters
# ======================
prewitt_x_kernel = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
prewitt_y_kernel = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])

prewitt_x = cv2.filter2D(img_gray, -1, prewitt_x_kernel)
prewitt_y = cv2.filter2D(img_gray, -1, prewitt_y_kernel)

# ======================
# 4. Laplacian Filter
# ======================
laplacian_kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
laplacian = cv2.filter2D(img_gray, -1, laplacian_kernel)

# ======================
# Plotting results
# ======================
titles = ["Original (Gray)", "Smoothing", 
          "Sobel X", "Sobel Y", 
          "Prewitt X", "Prewitt Y", 
          "Laplacian"]

images = [img_gray, smooth, 
          sobel_x, sobel_y, 
          prewitt_x, prewitt_y, 
          laplacian]

plt.figure(figsize=(12, 8))
for i in range(len(images)):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
