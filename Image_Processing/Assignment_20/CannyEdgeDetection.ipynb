implementing Canny Edge Detection (CED) algorithm in Python.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load and convert to grayscale
img = cv2.imread('art.png')  # Replace with your image path
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 2: Apply Gaussian Blur for noise reduction
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

# Step 3: Compute gradient using Sobel operator
sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)
gradient_direction = np.arctan2(sobely, sobelx)

# Convert gradient magnitude to uint8 for visualization
grad_mag_uint8 = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))

# Step 4: Apply Canny Edge Detection
edges = cv2.Canny(blurred, threshold1=100, threshold2=200)



# Step 5: Display results using Matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")


plt.subplot(1, 3, 2)
plt.imshow(grad_mag_uint8, cmap='gray')
plt.title("Gradient Magnitude")

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")

plt.tight_layout()
plt.show()
```

analyzing effects of different size Gaussian kernels

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

gray = cv2.imread("art.png" , 0)  # Load image in grayscale

b1 = cv2.GaussianBlur(gray , (3,3) ,0)
b2 = cv2.GaussianBlur(gray , (5,5,) , 0)
b3 = cv2.GaussianBlur(gray , (7,7) , 0)

edges1 = cv2.Canny(b1 , 100 , 200)
edges2 = cv2.Canny(b2 , 100 , 200)
edges3 = cv2.Canny(b3 , 100 , 200)

#without gaussion blur.......
edges_no_blur = cv2.Canny(gray, 100,200)

plt.figure()
plt.subplot(1,1,1)
plt.imshow(edges_no_blur , cmap = 'gray')
plt.title("No Gaussian Blur")
# plt.show()

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.imshow(gray , cmap = 'gray')
plt.title("Original Image")


plt.subplot(2,2,2)
plt.imshow(edges1 , cmap = 'gray')
plt.title("Gaussian Blur (3,3)")

plt.subplot(2,2,3)
plt.imshow(edges2 , cmap = 'gray')
plt.title("Gaussian Blur (5,5)")

plt.subplot(2,2,4)
plt.imshow(edges3 , cmap = 'gray')
plt.title("Gaussian Blur (7,7)")

plt.show()
plt.tight_layout()
```
